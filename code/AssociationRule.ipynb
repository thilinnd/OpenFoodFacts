```python
import pandas as pd
import warnings

# 1. Tắt cảnh báo đỏ để nhìn kết quả cho sạch
warnings.filterwarnings('ignore')

# 2. Đọc lại file đầy đủ (QUAN TRỌNG: Đã bỏ nrows=5)
# Lưu ý: Đảm bảo biến file_path vẫn đúng như bước trước
df = pd.read_csv(file_path, sep=None, engine='python', on_bad_lines='skip')

print(f"✅ Đã đọc lại toàn bộ dữ liệu. Tổng số dòng: {len(df)}")

# 3. Thực hiện lại bước Rời rạc hóa (như code trước nhưng trên dữ liệu full)
cols_interest = ['additives_n', 'nutriscore_grade', 'nova_group', 'general_category']
df_mining = df[cols_interest].copy()
df_mining.dropna(inplace=True)

# Phân nhóm additives_n
bins_additives = [-1, 0, 3, 100]
labels_additives = ['No Additives', 'Low Additives', 'High Additives']
df_mining['additives_cat'] = pd.cut(df_mining['additives_n'], bins=bins_additives, labels=labels_additives)

# Phân nhóm nova_group
df_mining['nova_cat'] = 'Nova_' + df_mining['nova_group'].astype(int).astype(str)

# Chọn cột cuối cùng
df_final = df_mining[['additives_cat', 'nova_cat', 'nutriscore_grade', 'general_category']]
```

```python

# 4. Tạo lại ma trận One-Hot
basket = pd.get_dummies(df_final, prefix_sep='=')
basket = basket.astype(bool)

print(f"✅ Kích thước ma trận MỚI: {basket.shape}")
# Bây giờ số dòng (số đầu tiên) phải lớn (ví dụ vài nghìn) chứ không phải là 5 nữa.
```

```python
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
```

```python
# THAM SỐ GỢI Ý
MIN_SUPPORT = 0.02  # Mẫu xuất hiện tối thiểu trong 2% sản phẩm

# Sử dụng Apriori (hoặc fpgrowth nếu dữ liệu quá lớn) để tìm tập phổ biến
frequent_itemsets = apriori(basket, min_support=MIN_SUPPORT, use_colnames=True)

# Thêm cột độ dài để dễ lọc (ví dụ chỉ quan tâm mẫu có từ 2 thuộc tính trở lên)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

print(f"✅ Tìm thấy {len(frequent_itemsets)} tập phổ biến.")
display(frequent_itemsets.sort_values(by='support', ascending=False).head())
```

```python
# THAM SỐ GỢI Ý
MIN_CONFIDENCE = 0.6 # Độ tin cậy tối thiểu 30%
MIN_LIFT = 1.0       # Lift > 1: Mối quan hệ tương quan dương (hỗ trợ nhau)

# Sinh luật
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=MIN_CONFIDENCE)

# Lọc các luật có Lift > 1 (để loại bỏ các luật ngẫu nhiên)
rules = rules[rules['lift'] > MIN_LIFT]

# Sắp xếp luật theo Lift giảm dần (những luật mạnh nhất lên đầu)
rules = rules.sort_values(by='lift', ascending=False)

print(f"✅ Tìm thấy {len(rules)} luật kết hợp.")
```

```python
# Lọc luật để trả lời câu hỏi: "Chế biến nhiều (Nova_4) có đi kèm dinh dưỡng kém (Nutriscore D hoặc E)?"

# Tìm các luật mà Vế trái (antecedents) chứa 'Nova_4'
# Và Vế phải (consequents) chứa Nutriscore D hoặc E
target_rules = rules[
    rules['antecedents'].apply(lambda x: 'nova_cat=Nova_4' in x) &
    rules['consequents'].apply(lambda x: 'nutriscore_grade=D' in x or 'nutriscore_grade=E' in x)
]

print("🔹 Các luật liên quan đến Thực phẩm siêu chế biến & Dinh dưỡng kém:")
# Chọn các cột quan trọng để hiển thị
cols_show = ['antecedents', 'consequents', 'support', 'confidence', 'lift']
display(target_rules[cols_show].head(10))
```

```python
# 1. Kiểm tra số lượng luật lọc được
print(f"🔍 Số luật thỏa mãn điều kiện lọc: {len(target_rules)}")

if len(target_rules) == 0:
    print("⚠️ Không tìm thấy luật chính xác 'Nova 4 -> Nutriscore D/E'.")
    print("👉 Gợi ý: Hãy xem Nova 4 thường đi kèm với cái gì khác (không ép buộc vế phải là Nutriscore):")

    # Lọc tất cả luật mà vế trái có Nova_4 (không quan tâm vế phải là gì)
    nova_rules = rules[rules['antecedents'].apply(lambda x: 'nova_cat=Nova_4' in x)]

    if len(nova_rules) > 0:
        display(nova_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10))
    else:
        print("❌ Cũng không có luật nào bắt đầu bằng Nova_4 luôn. Có thể Support của Nova_4 quá thấp hoặc bị lọc mất.")

else:
    # Nếu có kết quả thì hiển thị
    display(target_rules[cols_show])
```

```python
# Sử dụng lại biến 'basket' từ bước trước (ma trận True/False)
# min_support = 0.02: Chỉ lấy các tổ hợp xuất hiện trên ít nhất 2% tổng số sản phẩm
frequent_itemsets = apriori(basket, min_support=0.02, use_colnames=True)

# Thêm cột độ dài (length) để biết tổ hợp đó gồm mấy yếu tố
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

# --- LỌC VÀ HIỂN THỊ ---
# Chỉ lấy các tổ hợp có từ 2 yếu tố trở lên (để trả lời câu hỏi "đi kèm với nhau")
patterns = frequent_itemsets[frequent_itemsets['length'] >= 2].copy()

# Chuyển đổi format set -> string cho dễ đọc (bỏ chữ frozenset)
patterns['itemsets'] = patterns['itemsets'].apply(lambda x: ', '.join(list(x)))

# Sắp xếp: Cái nào phổ biến nhất lên đầu
patterns = patterns.sort_values(by='support', ascending=False)

print(f"✅ Tìm thấy {len(patterns)} mẫu kết hợp phổ biến (trên 2%).")
print("🔹 Top 15 tổ hợp thuộc tính xuất hiện nhiều nhất:")

# Hiển thị bảng kết quả
display(patterns[['itemsets', 'support']].head(15))
```

# NGÂN: nhờ Ngân vẽ phần FP-growth tree nhé

```python
#1 transaction = 1 sp
transactions = []

for _, row in basket.iterrows():
    items = [col for col, val in row.items() if val]
    transactions.append(items)

print("Ví dụ 5 transactions đầu:")
for t in transactions[:5]:
    print(t)
```

```python
#đếm tần suất item
from collections import Counter

item_counter = Counter()

for t in transactions:
    item_counter.update(t)

item_counter
```

```python
TOTAL = len(transactions)
MIN_SUPPORT = 0.02 * TOTAL

freq_items = {
    item: cnt for item, cnt in item_counter.items()
    if cnt >= MIN_SUPPORT
}

len(freq_items), freq_items
```

```python
# Header table đã sắp xếp giảm dần
freq_items_sorted = dict(
    sorted(freq_items.items(), key=lambda x: x[1], reverse=True)
)

ordered_transactions = []

for t in transactions:
    # Giữ lại item đạt min_support
    filtered = [i for i in t if i in freq_items_sorted]

    # Sắp xếp theo support toàn cục (FP-Growth chuẩn)
    filtered.sort(key=lambda i: freq_items_sorted[i], reverse=True)

    if filtered:
        ordered_transactions.append(filtered)

# Kiểm tra lại
ordered_transactions[:5]

```

```python
class FPNode:
    def __init__(self, item, count=1, parent=None):
        self.item = item              # tên item
        self.count = count            # số lần xuất hiện
        self.parent = parent          # node cha
        self.children = {}            # các node con
        self.next = None              # liên kết header table

    def increment(self, count=1):
        self.count += count
```

```python
def insert_transaction(transaction, node):
    """
    transaction: list item đã sắp xếp
    node: node hiện tại (bắt đầu từ root)
    """
    first_item = transaction[0]

    # Nếu node con đã tồn tại → tăng count
    if first_item in node.children:
        node.children[first_item].increment()
    else:
        # Tạo node mới
        new_node = FPNode(first_item, 1, node)
        node.children[first_item] = new_node

    # Đệ quy với phần còn lại
    if len(transaction) > 1:
        insert_transaction(transaction[1:], node.children[first_item])
```

```python
def build_fp_tree(transactions):
    root = FPNode("ROOT", count=0)

    for transaction in transactions:
        insert_transaction(transaction, root)

    return root
```

```python
def print_fp_tree(node, indent=0):
    print("  " * indent + f"{node.item}:{node.count}")
    for child in node.children.values():
        print_fp_tree(child, indent + 1)
```

```python
fp_root = build_fp_tree(ordered_transactions)

print_fp_tree(fp_root)
```

```python
from graphviz import Digraph

def draw_fp_tree(node, graph=None, parent_id=None):
    if graph is None:
        graph = Digraph(format="png")
        graph.attr(rankdir="TB")

    node_id = str(id(node))
    label = f"{node.item}\n{node.count}"
    graph.node(node_id, label=label, shape="box")

    if parent_id:
        graph.edge(parent_id, node_id)

    for child in node.children.values():
        draw_fp_tree(child, graph, node_id)

    return graph
```

```python
fp_root = build_fp_tree(ordered_transactions[:100])
fp_graph = draw_fp_tree(fp_root)

fp_graph
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def build_graph(node, G=None):
    if G is None:
        G = nx.DiGraph()

    for child in node.children.values():
        G.add_edge(f"{node.item}:{node.count}",
                   f"{child.item}:{child.count}")
        build_graph(child, G)

    return G

G = build_graph(fp_root)

plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=3000, font_size=8)
plt.show()
```

```python
def hierarchy_pos(G, root, width=1., vert_gap=0.2, vert_loc=0, xcenter=0.5):
    pos = {}

    def _hierarchy_pos(G, root, left, right, vert_loc):
        pos[root] = ((left + right) / 2, vert_loc)
        children = list(G.successors(root))
        if children:
            dx = (right - left) / len(children)
            nextx = left
            for child in children:
                _hierarchy_pos(G, child, nextx, nextx + dx, vert_loc - vert_gap)
                nextx += dx

    _hierarchy_pos(G, root, 0, width, vert_loc)
    return pos
```

```python
plt.figure(figsize=(14, 8))

root = "ROOT:0"
pos = hierarchy_pos(G, root)

nx.draw(
    G, pos,
    with_labels=True,
    node_size=3000,
    font_size=8,
    arrows=True,
)

plt.show()
```
